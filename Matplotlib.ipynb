# Matplotlib Kütüphanesi

 ## Gerekli Kütüphanelerin yüklenmesi

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Line Plot

```python
data = np.random.random(100)
#plot() çizgi grafiği çizmeye yarar. 
plt.plot(data)
#show() ile çizilen grafiği gösteririz.
plt.show()
```

```python
data = np.random.random(100)
#plot() çizgi grafiği çizmeye yarar. 
plt.plot(data, ls = ':')
#show() ile çizilen grafiği gösteririz.
plt.show()
```

```python
data = np.random.random(100)
#plot() çizgi grafiği çizmeye yarar. 
plt.plot(data, ls = 'None')
#show() ile çizilen grafiği gösteririz.
plt.show()
```

Matplotlib ile grafiğin rengini değiştirebiliriz.

```python
data = np.random.random(100)
#plot() çizgi grafiği çizmeye yarar. 
plt.plot(data, ls = ':', c = 'red') 
# c parametresine renk kodu vererek graiğin rengini değiştirebiliriz.
#show() ile çizilen grafiği gösteririz.
plt.show()
```

```python
data = np.random.random(100)
#plot() çizgi grafiği çizmeye yarar. 
plt.plot(data, ls = ':', c = '#4CAF50') 
# c parametresine renk kodu vererek graiğin rengini değiştirebiliriz.
#show() ile çizilen grafiği gösteririz.
plt.show()
```

linewidth parametresi ile grafiğin kalınlınlığını değiştirebiliriz.

```python
data = np.random.random(100)
#plot() çizgi grafiği çizmeye yarar. 
plt.plot(data, ls = ':', c = 'red', linewidth = '5') 
# c parametresine renk kodu vererek graiğin rengini değiştirebiliriz.
#show() ile çizilen grafiği gösteririz.
plt.show()
```

### Verileri tek grafikte gösterme

```python
sin_array = np.linspace(-np.pi, np.pi, 12)
out_sin = np.sin(sin_array)

cos = np.linspace(-(2*np.pi), 2*np.pi, 20)
out_cos = np.cos(cos)
plt.plot(out_cos)
plt.plot(out_sin)
plt.show()
```

```python
sin_array = np.linspace(-np.pi, np.pi, 12)
out_sin = np.sin(sin_array)

cos = np.linspace(-(2*np.pi), 2*np.pi, 20)
out_cos = np.cos(cos)
plt.plot(out_cos)
plt.plot(out_sin)
plt.title("Sin ve Cos Grafikleri") # Grafik Adı
plt.xlabel("Sin ve Cos Değerleri") # X ekseni
plt.ylabel("-pi +pi değerleri") # Y ekse
plt.show()
```

## Subplot

```python
#plot 1:
sin_array = np.linspace(-np.pi, np.pi, 12)
out_sin = np.sin(sin_array)
plt.subplot(2, 1, 1)
#Marker x ve y değerlerinin kesişimlerini işaretledik
plt.plot(sin_array,out_sin, c = 'g', ls = '--', marker = 'o')

#plot 2:
cos = np.linspace(-(2*np.pi), 2*np.pi, 20)
out_cos = np.cos(cos)

plt.subplot(2, 1, 2)
plt.plot(cos,out_cos, c = 'r', ls = ':', marker = '*')

plt.show()
```

## Scatter Plot

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, color = colors)
plt.colorbar()
plt.show()
```

## Bar Plot

```python
x = np.array(["Pazartesi", "Salı", "Çarşamba", "Perşembe", "Cuma", "Cumartesi", "Pazar"])
y = np.array([3, 8, 1, 10, 5, 8, 9])
plt.figure(figsize=(10,5)) # Tabloyu bütüp iç içe geçmeyi engellemek için kullandım.
plt.title("Günlere Göre Satış Grafiği")
plt.xlabel("Günler")
plt.ylabel("Satış Miktarı(bin tl)")
plt.bar(x,y)
plt.show()
```

## Histogram
Histogram, gruplandırılmış bir veri dağılımının sütun grafiğiyle gösterimidir. Diğer bir ifadeyle, tekrarlı sayılardan oluşan verilerin, uygulanan işlemlerden sonra önce tabloya, tablodan yararlanarak grafiğe aktarılması, yani veri gruplarının grafiğinin dikdörtgen sütunlar halinde gösterilmesidir.

```python
data = np.random.normal(170, 10, 250)

plt.hist(data)
plt.show()
```
